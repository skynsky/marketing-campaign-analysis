```python
df = pd.read_csv("data kotor.csv")
df.head()
```

```python
print(df.columns)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
df.describe()
```

```python
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outlier_income = df[(df['Income'] < lower) | (df['Income'] > upper)]

print("Jumlah Outlier:", len(outlier_income))
outlier_income[['ID','Income']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(y=df['Income'])
plt.title('Boxplot Income')
plt.ylabel('Income')
plt.show()
```

```python
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df = df[(df['Income'] >= lower) & (df['Income'] <= upper)]
```

```python
print(df.shape)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.boxplot(y=df['Income'])
plt.title('Boxplot Income Setelah Penghapusan Outlier')
plt.show()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outlier = df[(df['Income'] < lower) | (df['Income'] > upper)]

print("Jumlah Outlier:", len(outlier))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

corr_matrix = df.corr(numeric_only=True)

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f'
)

plt.title('Heatmap Korelasi Variabel Numerik')
plt.show()
```
